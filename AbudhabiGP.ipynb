```python
import fastf1
session = fastf1.get_session(2025, 'Saudi Arabia', 'Q')
session.load()
```

```python

session.load()
```

```python
fast_leclerc = session.laps.pick_drivers('1').pick_fastest()
print(fast_leclerc)
```

```python
print(session.laps.pick_drivers('1').get_car_data())
```

```python
from matplotlib import pyplot as plt
import fastf1.plotting
fastf1.plotting.setup_mpl(misc_mpl_mods=False, color_scheme='fastf1')
ver = session.laps.pick_drivers('1').pick_fastest()
vercar_data = ver.get_car_data()
t = vercar_data['Time']
vcar = vercar_data['Speed']
print(max(vcar))
```

```python
osc = session.laps.pick_drivers('PIA').pick_fastest()
osccar_data = osc.get_car_data()
tosc = osccar_data['Time']
ocar = osccar_data['Speed']
print(max(ocar))
```

```python

plt.plot(t, vcar, label='ver',color='blue')
plt.plot(tosc, ocar, label='osc',color='orange')
plt.xlabel('Time')
plt.ylabel('Speed [Km/h]')
plt.title('ver and osc')
plt.legend()
plt.grid(True)
plt.show()
```

```python
lec = session.laps.pick_drivers('LEC').pick_fastest()
leccar_data = lec.get_car_data()
tlec = leccar_data['Time']
lcar = leccar_data['Speed']
print(max(lcar))
```

```python
plt.plot(t, vcar, label='ver',color='blue')
plt.plot(tlec, lcar, label='lec',color='red')
plt.xlabel('Time')
plt.ylabel('Speed [Km/h]')
plt.title('max and leclerc')
plt.legend()
plt.grid(True)
plt.show()
```

```python
osc = sessionr.laps.pick_drivers('PIA').get_car_data()
lec = sessionr.laps.pick_drivers('LEC').get_car_data()
ver = sessionr.laps.pick_drivers('VER').get_car_data()
carl = sessionr.laps.pick_drivers('SAI').get_car_data()
rus = sessionr.laps.pick_drivers('RUS').get_car_data()

lcar = lec['Speed']
ocar = osc['Speed']
vcar = ver['Speed']
ccar = carl['Speed']
rcar = rus['Speed']
print(f"max speed of Ferrari  : {max(lcar)}")
print(f"max speed of Mclaren  : {max(ocar)}")
print(f"max speed of Redbull  : {max(vcar)}")
print(f"max speed of Williams  : {max(ccar)}")
print(f"max speed of Merc  : {max(rcar)}")
```
